```python
def zad4():
  n = int(input()) #вносим число N - количество чисел
  list = []
  for i in range(1,n+1):
    a = float(input()) #вносим вещественные числа
    A = round(a**i,4)
    i +=1
    list.append(A) #заполняем список
  print(list)
```
