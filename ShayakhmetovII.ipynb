```python
def xtrue(xxx): #Проверяет правильно ли введены координаты у X
  if (xxx[0] == xxx[1] and xxx[0]!=xxx[2] and xxx[1]!=xxx[2]):
    return True
  elif (xxx[0] == xxx[2] and xxx[0]!=xxx[1] and xxx[2]!=xxx[1]):
    return True
  elif (xxx[1] == xxx[2] and xxx[1]!=xxx[0] and xxx[2]!=xxx[0]):
    return True
  else:
        print("Введены неверные данные")
        return False

def ytrue(yyy): #Проверяет правильно ли введены координаты у Y
  if (yyy[0] == yyy[1] and yyy[0]!=yyy[2] and yyy[1]!=yyy[2]):
    return True
  elif (yyy[0] == yyy[2] and yyy[0]!=yyy[1] and yyy[2]!=yyy[1]):
    return True
  elif (yyy[1] == yyy[2] and yyy[1]!=yyy[0] and yyy[2]!=yyy[0]):
    return True
  else:
        print("Введены неверные данные")
        return False

def xycheck():
  print("Введите значения X")
  while True:
    x1 = float(input())
    x2 = float(input())
    x3 = float(input())
    x123 = [x1, x2, x3]
    if xtrue(x123):break
  x4 = findcoordx(x123)
  print("Введите значения Y")
  while True:
    y1 = float(input())
    y2 = float(input())
    y3 = float(input())
    y123 = [y1, y2, y3]
    if ytrue(y123):break
  y4 = findcoordy(y123)
  print("Координаты X", x4, "координаты Y", y4)

def findcoordx(xxx): #считает X
  if xxx[1]==xxx[2]:
     x4=xxx[0]
  if xxx[0]==xxx[1]:
     x4=xxx[2]
  if xxx[0]==xxx[2]:
     x4=xxx[1]
  return int(x4)

def findcoordy(yyy): #считает Y
  if yyy[1]==yyy[2]:
     y4=yyy[0]
  if yyy[0]==yyy[1]:
     y4=yyy[2]
  if yyy[0]==yyy[2]:
     y4=yyy[1]
  return int(y4)



```

Задача 5 - на поиск возраста человека

```python
from datetime import datetime as dt
def bornbornborn():
  born_date_str = input("Введите дату рождения в формате (dd-mm-yyyy)")
  current_datetime = dt.now().date() #берет текущую дату
  born_date = dt.strptime(born_date_str, "%d-%m-%Y").date() #переводит введенную дату в формат даты
  age = abs(current_datetime-born_date).days #высчитывает разницу в днях
  ageyear=age//365 #считает полные года
  print("Вам" ,ageyear, "года")
```

```python
bornbornborn()
```
